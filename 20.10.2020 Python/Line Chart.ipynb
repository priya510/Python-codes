```python
week1=["mon","tue","wed","thu","fri","sat","sun"]
week2=["mon","tue","wed","thu","fri","sat","sun"]
```

```python
temp1=[27,28,29,28.5,28,28.5,29]
temp2=[26,27,26,25.5,25,25.5,25]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(week1,temp1,color="red",label="Week1")
plt.plot(week2,temp2,color="blue",label="Week2")
plt.xlabel("Week")
plt.ylabel("Temp")
plt.title("Week 1 vs Week 2")
plt.legend()
```
