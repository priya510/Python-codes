```python
import pandas as pd
```

```python
#Display all student details.
df=pd.read_csv("studentdetails.csv")
df
```

```python
#Create new column full_name.
df["Full_Name"] = df["first_name"] + " "+ df["last_name"]
df
```

```python
#Add new column total_marks
df["Total_Marks"] =  df["subject1"] + df["subject2"] + df["subject3"]
df
```

```python
#Add new column percentage
df["Percentage"] = df["Total_Marks"]/300*100
df
```

```python
#Create a new column status (Pass or Fail)
df["Status"] = df["Percentage"].apply(lambda per:"Pass" if per>=50 else "Fail")
df
```

```python
#Add a grade column
def grade(per):
    if(per>=75):
        return "A"
    elif (per>=65):
        return "B"
    elif(per>=55):
        return "C"
    elif(per>=45):
        return "D"
    elif(per>=35):
        return "E"
    else:
        return "Fail"
    
    
df["grade"]=df["Percentage"].apply(grade)
df
```
